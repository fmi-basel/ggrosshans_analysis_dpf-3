# Wild type A

```python
df_input = pd.read_csv("../../01_small_RNA_seq_15_C/results/wild_type_A/htseq/alignment.sorted.filtered.unique_mappers.count_sequences/counts.tsv", header=0, sep="\t")
```

```python
df = pd.DataFrame(np.repeat(df_input.Name, df_input.counts))
df.columns = ["Name"]
df.index = np.arange(len(df))
df["read_length"] = df["Name"].str.len()
df["first_base"] = df["Name"].str[0]
plt.figure(figsize=(20,10))
sns.countplot(x="read_length", hue="first_base", hue_order=["A", "C", "G", "T"], data=df)
plt.savefig("WT_A.pdf")
```
