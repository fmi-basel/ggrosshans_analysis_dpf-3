```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Global options

sns.set_style("white")
sns.set(font_scale=2, style="ticks")

height=15
aspect=1
x_jitter=0
y_jitter=0

# xlim=(0,10)
ylim=(-3, 3)
alpha=1
marker='.'
s=1
legend=True
legend_out=False
linewidth=0
x_jitter=0
y_jitter=0
plt.rcParams["figure.figsize"] = (20,20)
```

```python
WT__Dpf_3_null_21U_RNA = pd.read_csv("../../01_small_RNA_seq_15_C/results/filter_sequences/DE_sequences_1_alignment_per_read__21T__WT__Dpf_3_null/DE_edgeR/final_table.tsv",
                                     header=0,
                                     sep="\t")
```

```python
WT__Dpf_3_null_21U_RNA["Significant"] = "No"
WT__Dpf_3_null_21U_RNA.loc[(WT__Dpf_3_null_21U_RNA["FDR"]<0.05) & ((WT__Dpf_3_null_21U_RNA["logFC"])<0), "Significant"] = "FDR<0.05\nlogFC<0"

sns_plot = sns.lmplot(x=f"logCPM",
                      y=f"logFC",
                      data=WT__Dpf_3_null_21U_RNA,
                      fit_reg=False,
                      hue=f"Significant",
                      hue_order=["No", "FDR<0.05\nlogFC<0"],
                      palette=["black", "red"],
                      scatter_kws={'alpha':alpha},
                      height=height,
                      aspect=aspect,
                      x_jitter=x_jitter,
                      y_jitter=y_jitter,
                      markers=marker,
                      legend=legend,
                      legend_out = legend_out)
sns_plot = plt.gca()
sns_plot.set_title(f"Dpf-3 null over WT - 21U RNA")
sns_plot.spines['right'].set_visible(True)
sns_plot.spines['top'].set_visible(True)
sns_plot.set(ylim=ylim)


plt.savefig("Dpf-3_null_over_WT_21U_RNA.pdf")

WT__Dpf_3_null_21U_RNA["Significant"].value_counts()
```
